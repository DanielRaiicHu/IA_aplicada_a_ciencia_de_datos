#### **Como crear un buen prompt con chat gpt**

* Una de las habilidades más importantes para la interacción con IAs generativas como ChatGPT es la capacidad de crear buenos prompts, guiando la herramienta de acuerdo con tu conocimiento sobre el negocio. La creación de un prompt eficaz en ChatGPT es fundamental para obtener resultados precisos y relevantes para la resolución de problemas.

* En el contexto de un proyecto de análisis de datos, como el que estamos trabajando aquí, un prompt bien elaborado puede ser la clave para obtener insights valiosos. A continuación, se presentan algunas buenas prácticas y ejemplos, teniendo en cuenta un proyecto simple de Ciencia de Datos:

**Sé específico y directo:**

* Explica directamente lo que deseas que ChatGPT ejecute. Esto ayuda a la IA a entender mejor tu necesidad, proporcionando respuestas más precisas.

* Ejemplo: “Realiza un análisis exploratorio de datos en la base ‘zoop_ventas.csv' usando las bibliotecas de manipulación de datos de Python. Presenta un resumen estadístico de las ventas por región”.

**Proporciona contexto cuando sea necesario:**

* Informa a ChatGPT del contexto del proyecto, incluyendo datos, objetivos y restricciones. Esto puede mejorar significativamente la calidad de las respuestas. Sin embargo, evita información innecesaria que pueda confundir a la IA.

* Ejemplo: "Estoy trabajando en un proyecto de análisis de ventas de una tienda minorista. La base de datos contiene información de ventas diarias por categoría de producto y región en 2023. Ayúdame a calcular la suma de las ventas de 'Electrónicos' por trimestre".

**Especifica el formato de la respuesta:**

* Indica el formato deseado para la respuesta, como gráficos, tablas o explicaciones detalladas.

* Ejemplo: "Crea un gráfico de líneas que muestre la evolución de las ventas a lo largo del último año, segmentado por categoría de productos".

**Indica el nivel de detalle deseado:**

* Deja claro el nivel de detalle o profundidad de la explicación que estás buscando. Esto ayuda a la IA a ajustar la respuesta a tu necesidad de comprensión.

* Ejemplo: "¿Podrías proporcionarme un ejemplo comentado de cómo usar la función groupby() de pandas para agrupar datos por categoría y calcular la suma total de ventas para cada grupo?"

**Usa términos técnicos correctos:**

* Pon atención a la terminología específica relacionada con tu área de interés. Esto asegura que la IA comprenda el dominio de tu pregunta y responda de forma adecuada.

* Ejemplo: “¿Cómo implemento un modelo de regresión lineal en Python usando scikit-learn para predecir ventas futuras con base en gastos de publicidad?”

**Divide tareas complejas en etapas simples:**

*Si el proyecto involucra varias etapas, divídelo en partes más pequeñas e instruye a ChatGPT paso a paso. Esto facilita que la IA te ayude de manera eficaz.

* Ejemplo:

* "¿Cómo puedo manejar valores faltantes en mi conjunto de datos usando pandas?"

* "¿Cuál es la mejor manera de visualizar la distribución de edades en mi conjunto de datos usando matplotlib?"

* “¿Podrías ajustar el visual de la distribución de edades colocando las etiquetas y labels con tamaño 12 y color #CCCCCC?”

**Mantente abierto a interacciones:**

* Esté siempre preparado(a) para realizar ajustes en tus prompts basados en las respuestas recibidas. La interacción puede ayudar a refinar las respuestas y obtener exactamente lo que necesitas.

* Ejemplo: “Esta respuesta puede mejorar. ¿Qué tal si redondeamos los valores a dos decimales y mostramos en el gráfico solo los valores mayores a 1 millón (1e6)?”

* Estos consejos son fundamentales en la construcción de prompts bien instruidos, como los que encontrarás a lo largo de este curso. Observa cómo se describen los pasos y acciones en los prompts del curso y cómo puedes aplicarlos en tu proyecto utilizando ChatGPT.

* Asegúrate de comunicar tus intenciones de forma precisa, dividir tareas complejas en etapas manejables y proporcionar ejemplos cuando sea apropiado. Esto te ayudará a obtener respuestas más relevantes y útiles para tu proyecto, haciendo la interacción con el chat más eficaz.

> ###### <strong><font color="7b71ff" size =6>Prompt:</font></strong>

Ahora vamos a unir las dos bases de nuestro proyecto en una sola utilizando como llave la columna "ID_compra". Además, organizaremos las columnas en la siguiente secuencia:

* ID_compra
* fecha
* horario
* categoria
* precio_unitario
* cantidad
* envio
* metodo_pago
* ID_cliente
* edad
* sexo_biologico
* ciudad
* estado
* region
* cashback
* nota

Construye un código para ejecutar esta etapa usando Python y la biblioteca Pandas.


````python
# --- 1. Unir los DataFrames por la columna 'ID_compra' ---
# Realiza un merge interno para mantener solo los registros que existen en ambas bases
df = pd.merge(ventas, clientes, on='ID_compra', how='inner')

# --- 2. Reorganizar las columnas en el orden solicitado ---
orden_columnas = [
    'ID_compra',
    'fecha',
    'horario',               # Asumiendo que en el DataFrame original se llama 'hora', no 'horario'
    'categoria',
    'precio_unitario',
    'cantidad',
    'envio',
    'metodo_pago',
    'ID_cliente',
    'edad',
    'sexo_biologico',
    'ciudad',
    'estado',
    'region',
    'cashback',
    'nota'
]

# Renombrar columnas si es necesario, para que coincidan con el orden solicitado
df = df.rename(columns={
    'hora': 'horario',
    'categoria': 'categoria',
    'precio unitario': 'precio_unitario',
    'método de pago': 'metodo_pago',
    'sexo': 'sexo_biologico'
})

# Aplicar el orden de columnas deseado
df = df[orden_columnas]

# Mostrar las primeras filas del DataFrame unificado y ordenado
df.head()
````

#### **2da Pregunta:** ¿Cuál fue la facturación por categoria em 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> ##### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear una nueva columna llamada 'facturacion' con el cálculo: (precio_unitario * cantidad) + envio
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# 2. Agrupar la facturación total por categoría de producto
facturacion_por_categoria = df.groupby('categoria')['facturacion'].sum().reset_index()

# 3. Ordenar la tabla resultante de menor a mayor facturación (para barras horizontales)
facturacion_por_categoria = facturacion_por_categoria.sort_values(by='facturacion', ascending=False)

# 4. Crear un gráfico de barras horizontales usando seaborn
plt.figure(figsize=(10, 6))
sns.barplot(
    data=facturacion_por_categoria,
    x='facturacion',
    y='categoria',
    palette='viridis'
)

# 5. Personalizar el gráfico con título y etiquetas
plt.title('Facturación total por categoría', fontsize=14)
plt.xlabel('Facturación total')
plt.ylabel('Categoría de producto')
plt.tight_layout()

# 6. Mostrar el gráfico
plt.show()
````

```
meses = {
```

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear columna 'facturacion' si aún no existe (solo si no se ha creado previamente)
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# 2. Extraer el año y mes de la columna 'fecha' en formato YYYY-MM
df['año_mes'] = df['fecha'].dt.to_period('M').astype(str)

# 3. Agrupar por año y mes, sumando la facturación mensual
ventas_mensuales = df.groupby('año_mes')['facturacion'].sum().reset_index()

# 4. Convertir la columna 'año_mes' a datetime para ordenarla correctamente en el gráfico
ventas_mensuales['año_mes'] = pd.to_datetime(ventas_mensuales['año_mes'])

# 5. Crear una nueva columna 'mes' con el nombre traducido del mes en español
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}

ventas_mensuales['mes'] = ventas_mensuales['año_mes'].dt.month_name().map(meses)

# 6. Crear gráfico de líneas de la facturación mensual
plt.figure(figsize=(10, 5))
sns.lineplot(data=ventas_mensuales, x='mes', y='facturacion', marker='o')

# 7. Personalizar el gráfico
plt.title('Ventas mensuales totales de Zoop', fontsize=14)
plt.xlabel('Mes')
plt.ylabel('Facturación total')
plt.tight_layout()

# 8. Mostrar gráfico
plt.show()
````

* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Crear columna de trimestre a partir de la columna 'fecha'
df['trimestre'] = df['fecha'].dt.to_period('Q').astype(str)

# 2. Agrupar los datos por trimestre y método de pago, y sumar la facturación
ventas_trimestrales = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum()
ventas_trimestrales
````

````python
# Convertir el resultado de la agrupación en una tabla con los trimestres como filas y los métodos de pago como columnas.
# Cada celda representa la facturación total de un método de pago en un trimestre específico.
# fill_value=0 asegura que, si no hubo ventas con cierto método en un trimestre, se registre como 0 en lugar de NaN.
ventas_trimestrales = ventas_trimestrales.unstack(fill_value=0)
ventas_trimestrales
````

````python
# 3. Crear el gráfico de barras apiladas
plt.figure(figsize=(10, 6))
ventas_trimestrales.plot(
    kind='bar',
    stacked=True, # Apilato
    figsize=(10, 6),
    colormap='Paired'
)

# 4. Personalizar el gráfico
plt.title('Método de pago por trimestre de 2023', fontsize=14)
plt.xlabel('Trimestre')
plt.ylabel('Facturación total')
plt.xticks(rotation=45)
plt.legend(title='Método de pago')
plt.tight_layout()

# 5. Mostrar gráfico
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear columna 'facturacion' si aún no existe (solo si no se ha creado previamente)
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# 2. Extraer el año y mes de la columna 'fecha' en formato YYYY-MM
df['año_mes'] = df['fecha'].dt.to_period('M').astype(str)

# 3. Agrupar por año y mes, sumando la facturación mensual
ventas_mensuales = df.groupby('año_mes')['facturacion'].sum().reset_index()

# 4. Convertir la columna 'año_mes' a datetime para ordenarla correctamente en el gráfico
ventas_mensuales['año_mes'] = pd.to_datetime(ventas_mensuales['año_mes'])

# 5. Crear una nueva columna 'mes' con el nombre traducido del mes en español
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}

ventas_mensuales['mes'] = ventas_mensuales['año_mes'].dt.month_name().map(meses)

# 6. Crear gráfico de líneas de la facturación mensual
plt.figure(figsize=(10, 5))
sns.lineplot(data=ventas_mensuales, x='mes', y='facturacion', marker='o')

# 7. Personalizar el gráfico
plt.title('Ventas mensuales totales de Zoop', fontsize=14)
plt.xlabel('Mes')
plt.ylabel('Facturación total')
plt.tight_layout()

# 8. Mostrar gráfico
plt.show()
````

````python
def facturacion_por_categoria():

    # Calcular la facturación
    df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

    # Agrupar por categoría
    facturacion_categoria = df.groupby('categoria')['facturacion'].sum().reset_index()

    # Ordenar de mayor a menor
    facturacion_categoria = facturacion_categoria.sort_values(by='facturacion', ascending=False)

    # Calcular porcentaje de la categoría con mayor facturación
    top_categoria = facturacion_categoria.iloc[0]['facturacion']
    total_categoria = facturacion_categoria['facturacion'].sum()
    porcentaje = (top_categoria / total_categoria) * 100

    # Crear figura
    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    ax.set_facecolor(GRIS_1)

    # Crear gráfico de barras horizontales
    sns.barplot(
        data=facturacion_categoria,
        x='facturacion',
        y='categoria',
        orient='h',
        color=AQUA_1,
        ax=ax
    )

    # Estilo del título
    ax.set_title('Facturación por categoría de producto en 2023', fontsize=18, color=GRIS_5)

    # Estilo de eje Y
    ax.set_ylabel('')
    ax.set_xlabel('')
    ax.tick_params(axis='x', bottom=False, labelbottom=False)  # Suprime eje x
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    # Suprimir bordes
    for side in ['top', 'left', 'bottom', 'right']:
        ax.spines[side].set_visible(False)



    # Mostrar los valores al final de cada barra
    for index, value in enumerate(facturacion_categoria['facturacion']):
        ax.text(
            value,
            index,
            f'MXN$ {value:,.0f}',
            va='center',
            ha='left',
            fontsize=12,
            color=AZUL_1
        )

    # Agregar comentario explicativo
    texto = (
        f"$\\bf{porcentaje:.2f}$% de las ventas corresponden únicamente a las ventas de Electrónica.\n"
        "Esto equivale a casi el doble en relación con todas las demás categorías\n"
        "presentes en el portafolio de Zoop."
    )
    ax.text(
        x=0.4e7,
        y=5,
        s=texto,
        fontsize=12,
        color=AZUL_3
    )

    # Obtener el objeto ejes actual
    ax = plt.gca()

    # Eliminar cuadrícula de fondo completamente
    ax.grid(False)

    # Mostrar gráfico
    plt.tight_layout()
    return plt
````

````python
def box_plot_edad_sexo():

    # Establecer estilo sin grilla por defecto
    sns.set_theme(style="white")

    # Crear figura
    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    ax.set_facecolor(GRIS_1)

    # Crear boxplot
    sns.boxplot(
        x='sexo_biologico',
        y='edad',
        data=df,
        palette=[AZUL_1, AQUA_1],
        ax=ax
    )

    # Título
    ax.set_title('Distribución de la Edad por Sexo Biológico de los clientes', fontsize=18, color=GRIS_5)

    # Estilo de ejes y etiquetas
    ax.set_xlabel('')  # Ocultar etiqueta del eje x
    ax.set_ylabel('Edad', fontsize=12, color=AZUL_1)
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    # Eliminar bordes
    for spine in ['top', 'bottom', 'left', 'right']:
        ax.spines[spine].set_visible(False)

    # Mostrar grilla detrás del gráfico
    ax.set_axisbelow(True)
    ax.yaxis.grid(True, linestyle='--', color=GRIS_3)

    # Cálculos para anotaciones
    grupo = df.groupby("sexo_biologico")["edad"]

    # Segunda categoría (índice 1)
    texto_1 = (
        f"Mínimo = {grupo.min().iloc[1]}\n"
        f"1º quartil = {grupo.quantile(0.25).iloc[1]}\n"
        f"2º quartil = {grupo.median().iloc[1]}\n"
        f"3º quartil = {grupo.quantile(0.75).iloc[1]}\n"
        f"Máximo = {grupo.max().iloc[1]}"
    )

    ax.annotate(
        texto_1,
        xy=(-0.1, 45),
        xytext=(-0.4, 50),
        fontsize=12,
        color=AZUL_1,
        bbox=dict(boxstyle="round,pad=0.5", facecolor=BLANCO, edgecolor=GRIS_3),
        arrowprops=dict(arrowstyle='->', color=AZUL_1)
    )

    # Primera categoría (índice 0)
    texto_0 = (
        f"Mínimo = {grupo.min().iloc[0]}\n"
        f"1º quartil = {grupo.quantile(0.25).iloc[0]}\n"
        f"2º quartil = {grupo.median().iloc[0]}\n"
        f"3º quartil = {grupo.quantile(0.75).iloc[0]}\n"
        f"Máximo = {grupo.max().iloc[0]}"
    )

    ax.annotate(
        texto_0,
        xy=(0.9, 45),
        xytext=(0.6, 50),
        fontsize=12,
        color=AZUL_1,
        bbox=dict(boxstyle="round,pad=0.5", facecolor=BLANCO, edgecolor=GRIS_3),
        arrowprops=dict(arrowstyle='->', color=AZUL_1)
    )

    # Mostrar gráfico
    plt.tight_layout()
    return plt
````
